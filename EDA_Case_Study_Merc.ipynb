```python
# Heatmap
plt.figure(figsize=(9, 5)) # Setting the figuresize (in inches)

sns.heatmap(df.corr(), annot=True)
```

```python
# Getting the proportion of values
df['fuel_ID'].value_counts()
```

```python
# Step 10: Data Visualization
# Histogram for car prices
plt.figure(figsize = (10,5))
sns.histplot(df['price'], bins = 20, kde = True)
plt.title('Histogram of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Step 10: Data Visualization
# Histogram for car mileage
plt.figure(figsize = (10,5))
sns.histplot(df['mileage'], bins = 20, kde = True)
plt.title('Histogram of Car Mileage')
plt.xlabel('Mileage')
plt.ylabel('Frequency')
plt.show()
```
