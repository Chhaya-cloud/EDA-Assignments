```python
# Plot for mileage across years
plt.figure(figsize=(21,9))
sns.boxplot(x='year', y='mileage', data=df, palette='Set3')
plt.title('Mileage across years')
plt.xlabel('Year')
plt.ylabel('Mileage')
plt.show()
```

```python

# Pie chart for Fuel_ID
plt.figure(figsize=(10,10))
plt.pie(df['fuel_ID'].value_counts(),labels=df['fuel_ID'].value_counts().index, autopct='%1.2f%%',startangle=140)
plt.title('Fuel_ID Distribution')
plt.show()


```

```python
# Create a Pie chart for transmitting_ID denoting the compostion
plt.figure(figsize=(10,10))
plt.pie(df['transmission_ID'].value_counts(),labels=df['fuel_ID'].value_counts().index, autopct='%1.2f%%',startangle=140)
plt.title('transmission_ID Distribution')
plt.show()
```

```python
# Correlation Analysis
cm = df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(cm, annot=True)
plt.show()
```

```python
#-1 indicates perfect negative correlationship
#1 indicates perfect positive correlation
```
