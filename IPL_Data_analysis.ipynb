Who has the highest strike rate?

```python
# Calculate strike rate for each player (strike rate = (runs / balls faced) * 100)
strike_rate = df1.groupby('batsman').apply(lambda x: (x['batsman_runs'].sum() / len(x)) * 100)
highest_strike_rate = strike_rate.idxmax(), strike_rate.max()
print(f"Player with the highest strike rate: {highest_strike_rate[0]} with a strike rate of {highest_strike_rate[1]:.2f}")
```

Who is the leading wicket-taker?

What is the count of runs scored from boundaries in each season?

```python
# Filter for boundaries (fours and sixes) and sum runs by season
boundary_runs = df1[df1['batsman_runs'].isin([4, 6])].groupby('inning')['batsman_runs'].sum()
print("Runs scored from boundaries in each season:")
print(boundary_runs)
```

What is the run contribution from boundaries in each season?

```python
# Calculate total runs and boundary runs, then percentage contribution by season
total_runs = df1.groupby('inning')['batsman_runs'].sum()
boundary_runs = df1[df1['batsman_runs'].isin([4, 6])].groupby('inning')['batsman_runs'].sum()
boundary_contribution = (boundary_runs / total_runs) * 100
print("Run contribution from boundaries in each season (%):")
print(boundary_contribution)
```

Which team has scored the most runs in the first 6 overs?

```python
# Filter for deliveries in the first 6 overs and sum runs by team
powerplay_runs = df1[df1['over'] <= 6].groupby('batting_team')['batsman_runs'].sum()
most_powerplay_runs = powerplay_runs.idxmax(), powerplay_runs.max()
print(f"Team with most runs in first 6 overs: {most_powerplay_runs[0]} with {most_powerplay_runs[1]} runs")
```

Which team has scored the most runs in the last 4 overs?

```python
# Filter for deliveries in the last 4 overs and sum runs by team
death_overs_runs = df1[df1['over'] > 16].groupby('batting_team')['batsman_runs'].sum()
most_death_overs_runs = death_overs_runs.idxmax(), death_overs_runs.max()
print(f"Team with most runs in last 4 overs: {most_death_overs_runs[0]} with {most_death_overs_runs[1]} runs")
```

Which team has the best scoring run-rate in the first 6 overs?

```python
# Calculate runs and balls in first 6 overs by team and compute run-rate
powerplay_data = df1[df1['over'] <= 6].groupby('batting_team').agg({'batsman_runs': 'sum', 'ball': 'count'})
powerplay_data['run_rate'] = (powerplay_data['batsman_runs'] / powerplay_data['ball']) * 6
best_powerplay_run_rate = powerplay_data['run_rate'].idxmax(), powerplay_data['run_rate'].max()
print(f"Team with best run-rate in first 6 overs: {best_powerplay_run_rate[0]} with a run-rate of {best_powerplay_run_rate[1]:.2f}")
```

Which team has the best scoring run-rate in the last 4 overs?

```python
# Calculate runs and balls in last 4 overs by team and compute run-rate
death_overs_data = df1[df1['over'] > 16].groupby('batting_team').agg({'batsman_runs': 'sum', 'ball': 'count'})
death_overs_data['run_rate'] = (death_overs_data['batsman_runs'] / death_overs_data['ball']) * 6
best_death_overs_run_rate = death_overs_data['run_rate'].idxmax(), death_overs_data['run_rate'].max()
print(f"Team with best run-rate in last 4 overs: {best_death_overs_run_rate[0]} with a run-rate of {best_death_overs_run_rate[1]:.2f}")
```
