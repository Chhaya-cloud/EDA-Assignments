KDE stands for kernel density estimation
Bins are vertical rectangles

```python
# Setting the figure size
fig, ax = plt.subplots(3,1, figsize=(10, 5))


# Histogram for number of passengers
sns.histplot(df['num_passengers'], bins = 9, kde = False,  ax=ax[0])
ax[0].set_title('Histogram of Number of passengers')
ax[0].set_xlabel('Number of passengers')
ax[0].set_ylabel('Frequency')


# Histogram for purchase lead
sns.histplot(df['purchase_lead'], bins = 9, kde = False,  ax=ax[1])
ax[1].set_title('Histogram of Purchase Lead')
ax[1].set_xlabel('Purchase Lead')
ax[1].set_ylabel('Frequency')


# Histogram for flight dusration
sns.histplot(df['flight_duration'], bins = 9, kde = False,  ax=ax[2])
ax[2].set_title('Histogram of Flight Duration')
ax[2].set_xlabel('Flight Duration')
ax[2].set_ylabel('Frequency')


# Display all plots
plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 7))
cm=df.corr(numeric_only=True)

# Plot
sns.heatmap(cm, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Majority of the variables seems to be almost uncorrelated with respect to each of it.
```
